```python
import tkinter as tk
import tkinter.filedialog as fd
import pandas as pd
from pandas.errors import EmptyDataError 
import os

root = tk.Tk()
root.withdraw()
root.lift()
filez = fd.askopenfilenames(parent=root, title='Choose a file')
filepath = fd.askdirectory()+"/"

df_ampli='a'
df_flowcurve='a'
df_nrelax2='a'
df_nrelax3='a'
df_nrelax4='a'
df_grelax2='a'
df_grelax3='a'
df_grelax4='a'


names=['Amplitude Sweep',"Flow curve","N relax 1","N Relax 2","N relax 3","G relax 1","G relax 2","G relax 3"]

for file in filez:
    if os.path.getsize(file)<10:
        continue
    df= pd.read_csv(file,encoding='utf-16',delimiter='\t',skiprows=[0],skip_blank_lines=True).iloc[:,1:]
    df.at[0]=file.split("/")[-1].split(".")[0]
    print("G relaxation" in file and "Interval 3" in file)
    

    if "Amplitude" in file and "Interval 2" in file: 
        df=df[['Shear Strain','Storage Modulus','Loss Modulus']]
        if type(df_ampli)==str:
            df_ampli=df.copy()
        else : df_ampli=pd.concat([df_ampli,df],axis=1).reindex(df.index)
    
    if "Fast Flow Curve" in file and "Interval 3" in file: 
        df=df[['Shear Rate','Shear Stress','Viscosity']]
        if type(df_flowcurve)==str:
            df_flowcurve=df.copy()
        else : df_flowcurve=pd.concat([df_flowcurve,df],axis=1).reindex(df.index)
    
    if ("G relaxation" in file and "Interval 2" in file): 
        df=df[['Time','Storage Modulus','Loss Modulus']]
        print('trigger')
        if type(df_grelax2)==str:
            df_grelax2=df.copy()
        else : df_grelax2=pd.concat([df_grelax2,df],axis=1).reindex(df.index)
    if ("G relaxation" in file and "Interval 3" in file):
        print('trigger') 
        df=df[['Time','Storage Modulus','Loss Modulus']]
        if type(df_grelax3)==str:
            df_grelax3=df.copy()
            print(df_grelax3)
        else : df_grelax3=pd.concat([df_grelax3,df],axis=1).reindex(df.index)
    if ("G relaxation" in file and "Interval 4" in file): 
        df=df[['Time','Storage Modulus','Loss Modulus']]
        if type(df_grelax4)==str:
            df_grelax4=df.copy()
        else : df_grelax4=pd.concat([df_grelax4,df],axis=1).reindex(df.index)
    if ("Relaxation" in file and "Interval 2" in file and "G" not in file): 
        df=df[['Time','Viscosity']]
        if type(df_nrelax2)==str:
            df_nrelax2=df.copy()
        else : df_nrelax2=pd.concat([df_nrelax2,df],axis=1).reindex(df.index)
    if ("Relaxation" in file and "Interval 3" in file and "G" not in file): 
        df=df[['Time','Viscosity']]
        if type(df_nrelax3)==str:
            df_nrelax3=df.copy()
        else : df_nrelax3=pd.concat([df_nrelax3,df],axis=1).reindex(df.index)
    if ("Relaxation" in file and "Interval 4" in file and "G" not in file): 
        df=df[['Time','Viscosity']]
        if type(df_nrelax4)==str:
            df_nrelax4=df.copy()
        else : df_nrelax4=pd.concat([df_nrelax4,df],axis=1).reindex(df.index)

list=[df_ampli,df_flowcurve,df_nrelax2,df_nrelax3,df_nrelax4,df_grelax2,df_grelax3,df_grelax4]
print('*********')
print(df_grelax3)
print(type(df_grelax3))
print('*********')

for df,name in zip(list,names):
    print(df)
    if type(df) is not str:
        print('done')
        df.to_csv(filepath+name+".csv",encoding='utf-16')

```
