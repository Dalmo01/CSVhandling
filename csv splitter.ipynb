```python
import tkinter as tk
from tkinter import filedialog

root = tk.Tk()
root.withdraw()
root.lift()
global write
write=False

f= filedialog.askopenfilename()
filepath = filedialog.askdirectory()+"/"
print(filepath)

with open(f,encoding="utf-16") as data:
    title=filepath
    for line in data:
        print(write)
        # if line in (None,''):
        #     write=False
        #     title=filepath

        if "Result" in line:
            result= f.split("/")[-1].split(".")[0] + line.split(":")[1]
            
        if "Interval and data points" in line:
            interval="Interval "+line.split("\t")[-2]+".csv"
            
            
            write=True 
        if write == True:
            filename=title+result+interval
            with open(filename.replace("\t","").replace("\n",""), "a+",encoding='utf-16') as output:
                if not('Result' in line or line==''):
                    output.write(line)
        


    
```
